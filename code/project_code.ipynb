```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
df
```

# **data preparation**

## data speration as x and y

```python
y=df['Profit']
y
```

```python
from os import X_OK
X=df.drop('Profit',axis=1)
X
```

# **dividing data into train set and test set(80% train and 20% test)**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

# **Regression models**

## **Linear Regression**

###**Training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

###**model prediction**

```python
y_train_pred_lr = lr.predict(X_train)
y_test_pred_lr = lr.predict(X_test)
```

###**evaluate model(regression matrics)**

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
lr_train_mae=mean_absolute_error(y_train, y_train_pred_lr)
lr_train_mse=mean_squared_error(y_train, y_train_pred_lr)
lr_train_rmse=np.sqrt(lr_train_mse)
lr_train_r2=r2_score(y_train, y_train_pred_lr)

lr_test_mae=mean_absolute_error(y_test, y_test_pred_lr)
lr_test_mse=mean_squared_error(y_test, y_test_pred_lr)
lr_test_rmse=np.sqrt(lr_test_mse)
lr_test_r2=r2_score(y_test, y_test_pred_lr)
```

```python
lr_train_results=pd.DataFrame(['Linear Regression',lr_train_mae,lr_train_mse,lr_train_rmse,lr_train_r2]).transpose()
lr_test_results=pd.DataFrame(['Linear Regression',lr_test_mae,lr_test_mse,lr_test_rmse,lr_test_r2]).transpose()
```

```python
lr_train_results.columns=['method','train MAE','train MSE','train RMSE','train R2']
lr_test_results.columns=['method','test MAE','test MSE','test RMSE','test R2']
```

```python
lr_train_results
```

```python
lr_test_results
```

```python
poly_train_results=pd.DataFrame(['Polynomial Regression',poly_train_mae,poly_train_mse,poly_train_rmse,poly_train_r2]).transpose()
poly_test_results=pd.DataFrame(['Polynomial Regression',poly_test_mae,poly_test_mse,poly_test_rmse,poly_test_r2]).transpose()
```

```python
poly_train_results.columns=['method','train MAE','train MSE','train RMSE','train R2']
poly_test_results.columns=['method','test MAE','test MSE','test RMSE','test R2']
```
